### OUTLIER HANDLING
- Important features like 'age,' 'height_cm,' 'weight_kg', 'value_eur,' 'wage_eur,' and 'release_clause_eur' are continuous and need outlier handling because they can have extreme values that might skew the clustering process.
- Other features like 'shooting,' 'dribbling,' etc., have values in the range of 0 to 100, representing specific skills, and do not require outlier treatment as they are already bounded within a meaningful range.
- Outlier handling for continuous features involves identifying and transforming extreme values to ensure a more robust and accurate clustering of football players based on their skill attributes.

```python
 Data= ['age','height_cm','weight_kg','value_eur','wage_eur','release_clause_eur']

plt.figure(figsize=(20,25)) # defining canvas size
n = 1 

for i in Data: 
    if n<=9:    # set the limit
        plt.subplot(3,3,n) 
        sns.boxplot(x=fifa[i]) # Plotting box plots 
        plt.xlabel(i,fontsize=20)  # assigning name to x-axis and font size is 20
    n+=1 # counter increment
plt.tight_layout()
plt.show() # used to hide the storage loction
```

- Here age,height_cm,weight_kg,overall,potential follows a normal distribution so we are empherical rule and remaining feature we use iqr

##### EMPHERICAL RULE OR 3 STANDARD DEVIATION RULE TO DETECT THE OUTLIER
- For age,height,weight,overall,potential we use 3 sigma rule
- From the boxplot we can see that the outliers are not > 5% so we can use mean imputation for the outliers

```python
Data= ['age','height_cm','weight_kg','overall','potential']
for i in Data: 
    upper_limit = fifa[i].mean() + 3*fifa[i].std()
    lower_limit = fifa[i].mean() - 3*fifa[i].std()
    fifa.loc[fifa[i] > upper_limit,i]=fifa[i].mean()
    fifa.loc[fifa[i] < lower_limit,i]=fifa[i].mean()
```

##### IQR for positively skewed curve
- For 'value_eur','wage_eur','release_clause_eur' we use IQR

```python
from scipy import stats
Data=['value_eur','wage_eur','release_clause_eur']
for i in Data:
    iqr = stats.iqr(fifa[i])
    Q1 = fifa[i].quantile(0.25)  # first quantile
    Q3 = fifa[i].quantile(0.75)  #third quantile
    # getting max & min limit
    min_limit = Q1 - 1.5*iqr
    max_limit = Q3 + 1.5*iqr
    # get the percentage
    print('percentage of outliers for ',i,':',len(fifa.loc[fifa[i] > max_limit])/18278*100)
```

- since the outliers for 'value_eur','wage_eur','release_clause_eur' are greater than 5% we cannot use imputation on the outliers

### 1. KMEANS Clustering

### Importing the required model evaluation matrix:
- The silhouette score, davies_bouldin_score,calinski_harabasz_score are the clustering evaluation metric used to assess the quality of clustering results.
- Silhouette Score: It measures how well each data point belongs to its cluster compared to other clusters. A higher score means the clusters are well-defined and not too close to each other.
- Davies-Bouldin Score: It assesses clustering by looking at how compact each cluster is and how separated they are from each other. A lower score suggests clear and separate clusters without much overlap.
- Calinski-Harabasz Score: It evaluates clustering by comparing how far the clusters are from each other to how far the points are within the clusters. A higher score means the clusters are well-defined and distinct from each other

```python
from sklearn.metrics import silhouette_score,davies_bouldin_score,calinski_harabasz_score
```

```python
kmeans = KMeans(n_clusters=4,random_state=4)
kmeans.fit(principle_df)
Kmeans_silhou_4 = silhouette_score(principle_df,kmeans.labels_)
Kmeans_davies_4 = davies_bouldin_score(principle_df,kmeans.labels_)
Kmeans_calinski_4 = calinski_harabasz_score(principle_df,kmeans.labels_)
print("Silhouette Score:",Kmeans_silhou_4)
print('davies_score:',Kmeans_davies_4)
print('calinski:',Kmeans_calinski_4)
```

### Report on Challenges faced
1.Missing Data:
- Technique: Imputation
- Reason: For attributes with missing data, imputation techniques like mean, median,min are used to fill in the missing values, ensuring the integrity of the dataset for clustering.

2.High Dimensionality:
- Technique: Dimensionality Reduction 
- Reason: To handle high-dimensional data, techniques like Principal Component Analysis (PCA) is applied to reduce the number of attributes while preserving most of the variance.

3.Data Preprocessing:
- Technique: Labelencoding 
- Reason: The categorical attributes are manually labelencoded to converts them into numerical form suitable for clustering. 

4.Outliers:
- Technique: Z-score,IQR
- Reason:
- Outliers can distort clustering results
- techniques like Z-score or IQR (Interquartile Range) are used to detect the outliers.


5.Correlation:
- Highly correlation features are found using fifa.corr() function and features with correlation >=97 are dropped. 


6.Varying Data Scales:
- Technique: Minmaxsaclar 
- Reason: 
- Here MinMaxScaler scalar is used to sacle the data,
- Scaling attributes to a similar range prevents certain features from dominating the clustering process, ensuring a fair representation of all attributes.


7.Interpreting Clusters:
- Technique: Domain Knowledge 
- Reason: Domain knowledge is used to find the optimal number of cluster for fifa 20 dataset.

8.Determining Optimal Clustering Algorithm:
- Technique: Comparison and Evaluation
- Reason: Trying multiple clustering algorithms (e.g., KMeans, DBSCAN, Hierarchical) and evaluating their performance using metrics like Silhouette Score and Davies-Bouldin Score helps identify the best algorithm for the dataset.


By employing these techniques, the challenges faced in the FIFA20 dataset are effectively addressed, leading to meaningful clustering results that offer valuable insights into player skills and position classifications.
